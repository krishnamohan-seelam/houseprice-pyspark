```python
import pandas as pd
```

```python
train_file = "house-prices/train.csv"
```

```python
df = pd.read_csv(train_file,keep_default_na=False)
```

```python
df.head()
```

```python
from sdv.tabular import GaussianCopula
```

```python
model = GaussianCopula(primary_key='Id')
```

```python
model.fit(df)
```

```python
new_data = model.sample(num_rows=1500)
```

```python
new_data.head()
```

```python
from sdv.evaluation import evaluate
evaluate(new_data.drop('Id',axis=1), df.drop('Id',axis=1))
```

```python
new_data.info()
```

```python
new_data['Id'] = new_data['Id'] + 8000
```

```python
new_data.to_csv("house-prices/train_sdv.csv",index=False)
```
